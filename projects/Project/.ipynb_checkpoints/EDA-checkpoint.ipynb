```python
# Check data
df.info()
df["bmi"] = round(df["mass"]/(df["length"]/100)**2, 2)
df["factor"] = round(df["genetic"]/df["lifespan"], 2)
bmi_bins = [0,22,28,30, np.inf]
smoking_bins = [0, 5, 10, 15, 20, np.inf]
df["bmi bins"] = pd.cut(df['bmi'],bmi_bins, labels = ["underweight", "normal", "overweight", "obese"])
#df["bmi bins"] = pd.cut(df['bmi'],bmi_bins, labels = [1, 2, 3,4])

df["smoking_bins"] = pd.cut(df['smoking'],smoking_bins, labels = ["non smoking", "little smoking", "ten smoking", "15 smoking", "smoking"])
#df["smoking"] = pd.cut(df['smoking'],smoking_bins, labels = [),1, 2, 3,4])

df.head()
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(Q1)
print()
print(Q3)
print()
print(IQR)
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("factor")
plt.ylabel('number of people')
plt.xlabel('factor');
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
```
