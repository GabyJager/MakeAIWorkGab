```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(Q1)
print()
print(Q3)
print()
print(IQR)
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("lifespan")
plt.ylabel('number of people')
plt.xlabel('lifespan');
```
